```python
from textblob import TextBlob
import csv
import pandas as pd
import os
```

```python
path= '.'
files = os.listdir(path)
polarity = []
subjectivity = []
for i in files:
    if i[-3:] =='csv':
        
        with open(i) as f:
            s = f.read()

        blob=TextBlob(s)

        polarity.append(blob.sentiment.polarity)
        subjectivity.append(blob.sentiment.subjectivity)
```

```python
sentiment = pd.DataFrame({'Polarity':polarity,'Subjectivity':subjectivity})
```

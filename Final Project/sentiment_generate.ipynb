## The following is main code for generating sentiment

```python
def sentiment_analyze(text):
    blob= TextBlob(text)
    
    return blob.sentiment

```

```python
statement_dates = pd.read_csv('statement_dates.csv')
```

```python
def check_negate(word):
    ''' Function checks whether word is a negation word'''
    if word in negate_words:
        return True
    else:
        return False

def sentiment_check(pos_words, neg_words, text):
    # method from Loughran and Mcdonald (2011)
    '''Function checks each word in the text that is passed in, will check if match
        in either positive or negative word list. 
        
        Text must be a list of strings
        Will also check for negation to negate/reverse positive words
        
        Returns: word count, pos count, neg count'''
    
    pos = 0
    neg = 0

    
    text = [word for word in text if len(word)>1]
    
    for i in range(len(text)):
        if text[i] in neg_words:
            neg+=1
            print(text[i],'neg')
        elif text[i] in pos_words:
            if i>=3:
                
                if check_negate(text[i-3]) or check_negate(text[i-2]) or check_negate(text[i-1]):
                    neg+=1
                    print(text[i],'neg')
                else:
                    pos+=1
                    print(text[i], 'pos')
            elif i==2:
                if check_negate(text[i-2]) or check_negate(text[i-1]):
                    neg+=1
                    print(text[i],'neg')
                else:
                    pos+=1
                    print(text[i], 'pos')

            elif i==1:
                if check_negate(text[i-1]):
                    neg+=1
                    print(text[i],'neg')
                else: 
                    pos+=1
                    print(text[i], 'pos')
            elif i==0:
                pos+=1
                print(text[i], 'pos')
    
    lista = [word for word in text if word not in stopwords.words('english')]
    #text = ' '.join(lista)
    text = [word for word in text if len(word)>1]
                                
    word_count = len(text)
                                
    return {'Pos Words':pos,'Neg Words':neg, 'Word Count': word_count}
```

```python
polarity=[]
subjectivity=[]
lm_pos = []
lm_neg = []
cb_pos = []
cb_neg =[]
word_count = []
negate_words = ['NO', 'NOT', 'NONE', 'NEITHER', 'NEVER', 'NOBODY'] #From Loughran and Mcdonald (2011)


#LM dictionary read
fin_pos_words = pd.read_excel('LoughranMcDonald_SentimentWordLists_2018.xlsx',sheet_name='Positive',header=None)
fin_neg_words = pd.read_excel('LoughranMcDonald_SentimentWordLists_2018.xlsx',sheet_name='Negative',header=None)
fin_pos_words=fin_pos_words[0].values
fin_neg_words=fin_neg_words[0].values

#CB dicionary read

cb_words = pd.read_excel('ifdp1203-appendix.xlsx',sheet_name='1 FS Dictionary')
cb_pos_words = cb_words[cb_words['Positive']==1]['Word']
cb_pos_words = [word.upper() for word in cb_pos_words]
cb_neg_words = cb_words[cb_words['Negative']==1]['Word']
cb_neg_words = [word.upper() for word in cb_neg_words]



for date in statement_dates['Dates']:
    if date < 20190301:
        with open('.\scraped_data\\'+str(date)+'.csv') as f:
            print(str(date))
            text = f.read()
        
            text_sent=sentiment_analyze(text)
        
            polarity.append(text_sent.polarity)
            subjectivity.append(text_sent.subjectivity)
            
            word_list = [word.lemmatize().upper() for word in TextBlob(text).words]
            
            print('LM')
            lm_sent = sentiment_check(fin_pos_words,fin_neg_words,word_list)
            lm_pos.append(lm_sent['Pos Words'])
            lm_neg.append(lm_sent['Neg Words'])
            
            print('CB')
            cb_sent = sentiment_check(cb_pos_words,cb_neg_words,word_list)
            
            cb_pos.append(cb_sent['Pos Words'])
            cb_neg.append(cb_sent['Neg Words'])
            
            word_count.append(cb_sent['Word Count'])
            

            
```

```python
base_sentiment=pd.DataFrame({'polarity':polarity,'subjectivity':subjectivity,
                             'LM_POS': lm_pos,'LM_neg':lm_neg,
                             'CB_POS': cb_pos,'CB_neg':cb_neg,
                             'word_count':word_count})
```

```python
sentiment_frame = statement_dates.merge(base_sentiment,left_on=statement_dates.index,right_on=base_sentiment.index)
```

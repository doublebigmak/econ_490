```python
import statsmodels.api as sm
```

```python
merged['isFemale']= merged['SEX'].map({1:0,2:1})
```

```python
y = merged['HRLYEARN']
```

```python
x = merged[['AGE_12','PROV','CMA','isFemale','MARSTAT']]

reg1 = sm.OLS(endog=y, exog=x, missing='drop')
```

```python
results=reg1.fit()
results.summary()
```
